```python
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
# slice files location
slice_data_folder = r"e:\PerkLab\Data\Spine\Verdure\4_Slices"
```

```python
# get list of segmentation files
seg_files = glob.glob(os.path.join(slice_data_folder, "labels", "*/*.npy"), recursive=True)
print(len(seg_files))
```

```python
# get 10 random slices
np.random.seed(0)
rand_files = np.random.choice(seg_files, 10)
print(rand_files)
```

```python
# get matching image files
img_files = [f.replace("labels", "images").replace("segmentation", "ultrasound") for f in rand_files]
print(img_files)
```

```python
# display images and segmentations with overlay
fig, axes = plt.subplots(len(rand_files), 3, figsize=(8, 30))
for i, (img_file, seg_file) in enumerate(zip(img_files, rand_files)):
    img = np.load(img_file)
    seg = np.load(seg_file)
    axes[i, 0].imshow(img, cmap="gray")
    axes[i, 1].imshow(seg, cmap="gray")
    axes[i, 2].imshow(img, cmap="gray")
    axes[i, 2].imshow(seg, cmap="jet", alpha=0.5)
    axes[i, 0].axis("off")
    axes[i, 1].axis("off")
    axes[i, 2].axis("off")
```
